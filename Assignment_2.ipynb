14. Write a program that takes a number and checks whether it is positive or negative. If it is positive, check whether it is even or odd using a nested if.

15. Write a program that takes a student's marks and first checks if the marks are above or below 40 (pass/fail). If passed, use a nested if to decide the grade (A, B, or C).


```python
grade = int(input("Enter your grade: "))

if grade >= 40:
  if grade >= 80:
    print ("A")
  elif grade >= 70:
    print ("B")
  else:
    print ("C")
else:
  print ("Fail")
```

16. Write a program that takes a person's age. First check if the age is above 18 or not. If above 18, use a nested if to check if the age is above 60 as well (senior citizen).


```python
age = int(input("Enter your age: "))

if age > 18:
  if age > 60:
    print ("You're a senior citizen.")
  else:
    print ("You're not a senior citizen.")
```

18. Write a function named count_down(n) that uses a loop inside the function to print numbers from n down to 1.


```python
def count_down(n):
  for i in range (n, 0, -1):
    print (i)

count_down(4)
```

19. Write a function named sum_of_list(numbers) that takes a list and uses a loop inside the function to calculate and return the total sum of the list.


```python
def sum_of_list(numbers):
  total = 0
  for i in range (len(numbers)):
    total += numbers[i]

  return total

print (sum_of_list([1, 2, 3, 4, 5, 6]))
```
